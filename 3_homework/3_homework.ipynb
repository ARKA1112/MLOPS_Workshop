First do prefect project init, this creates

.prefectignore
prefect.yaml
.prefect/

/Start the prefect server

/One has to select the git option in order to deploy and pull from git repo, otherwise local and s3 deployments can also be considered.

/Next we will create a work pool by prefect work-pool create *name ...This can be anything from local subprocess to cron job or GCP, Azure job etc

/After creating the pool we can create a deployment by typing the following commmand 
prefect deploy file_location.py:main_flow --name "name of the deployment" --pool "name of the pool to be assigned..NB this pool needs to be created beforehand"


/After checking everything start the worker node by typing in a separate terminal preferably a new one -- prefect worker start -n "name of the worker", then deliver the work_pool_name that you created beforehand
